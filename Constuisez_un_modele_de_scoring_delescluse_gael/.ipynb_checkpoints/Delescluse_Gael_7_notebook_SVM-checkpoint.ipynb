## SVM
Les machines à vecteurs de support (SVM) sont un type d'algorithme d'apprentissage automatique utilisé pour la classification et la régression. Il cherche à trouver la meilleure séparation entre les groupes de données. Ils sont particulièrement utiles dans des situations où il y a un espace de caractéristiques complexe ou lorsque la séparation linéaire est difficile.

```python
import pandas as pd
import numpy as np
import time
import shap
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import make_scorer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
from sklearn.metrics import f1_score
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt
import seaborn as sns
```

Pour analyser la pertinence du modèle que nous sélectionnerons, nous devons créer une fonction qui pénalise les mauvaises prédiction et plus particulièrement les False Negative. Ce sont des prédictions du modèle qui prédise un 0 alors que nous attendons un 1. Cette prédiction est plus dangereuse pour la pertinence du modèle car si nous prédisons un remboursement probable d'un crédit et que cela n'est pas le cas, cela impacte directement la rentabilité de la banque, ce qui n'est pas souhaitable. Nous créons donc une fonction de scoring qui pénalise de 10 les False Negative et de 1 les False Positive.

```python
def custom_penalty(y_test, y_pred):
    y_test = y_test.to_numpy().flatten()
    penalty = 0.0
    sum_penalty = 0
    
    for yt, yp in zip(y_test, y_pred):
        if yt == 1 and yp == 0:
            penalty = penalty + 10
            sum_penalty = sum_penalty +1
        elif yt == 0 and yp == 1:
            penalty = penalty + 1
            sum_penalty = sum_penalty +1
            
    score = (penalty/len(y_test))*0.1#/ (sum_penalty + 1e-7) # Prevents zero division
    return 1 - score
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC  # pour la classification, ou SVR pour la régression
from sklearn.metrics import accuracy_score, classification_report
```

Nous récupérons les données d'entrainement et de test qui sont divisées dans différents fichier. Cette partie concerne le modèle de Machine Learning sur les données déséquilibrées et traitées avec une méthode d'Hybrid-resampling.

```python
X_train_hybrid = pd.read_csv('X_train_CSV_hybrid.csv')
X_test_hybrid = pd.read_csv('X_test_CSV_hybrid.csv')
y_train_hybrid = pd.read_csv('y_train_CSV_hybrid.csv')
y_test_hybrid = pd.read_csv('y_test_CSV_hybrid.csv')
```

Cette étape est uniquement destiné à accélerer le temps de traitement du modèle lors de l'utilisation de GridSearchCV, les résultats avec l'ensemble des variables sont sensiblement les mêmes. Ici nous limitons le nombre de variable permettant l'entraienement du modèle en sélectionnant les variables mis en valeur par l'étude de Features Importances des variables.

```python
selected_columns = ['EXT_SOURCE_3', 'EXT_SOURCE_2', 'NAME_FAMILY_STATUS_Married',\
                    'FLAG_OWN_CAR', 'DAYS_BIRTH', 'CODE_GENDER', 'FLAG_DOCUMENT_3',\
                    'DAYS_ID_PUBLISH', 'NAME_CONTRACT_TYPE',\
                    'WALLSMATERIAL_MODE_Panel', \
                    'FLAG_DOCUMENT_6',\
                    'FLAG_PHONE', 'FLAG_DOCUMENT_8',\
                    'WEEKDAY_APPR_PROCESS_START_MONDAY',\
                    'DAYS_EMPLOYED']
X_train_hybrid = X_train_hybrid[selected_columns].copy()
X_test_hybrid = X_test_hybrid[selected_columns].copy()
```

Nous effectuons une transformations de mise à l'échelle des caractéristiques des individus car certains modèles de Machine Learning sont sensibles à l'échelle des variables. Cette étape est également utile optimiser la performance de nos modèles.

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_hybrid)
X_test_scaled = scaler.transform(X_test_hybrid)
```

GridSearchCV permet de tester différents paramètres dans notre modèle de Machine Learning afin de trouver ceux qui maximisent le score du modèle. Ici GridSearchCV doit maximiser le score de notre fonction de pénalisation des résultats FALSE NEGATIVE.

```python
param_grid = {
    'C': [0.01, 0.1, 1, 10, 100]
}
svm = SVC(kernel='linear', cache_size=1000)

grid_search = GridSearchCV(svm, param_grid, scoring=custom_penalty, cv=5)
grid_search.fit(X_train_scaled, y_train_hybrid.values.ravel())
grid_search
```

```python
best_params = grid_search.best_params_
best_score = grid_search.best_score_
best_model = grid_search.best_estimator_
best_model_score = best_model.score(X_test_scaled, y_test_hybrid)
print(best_params)
print(best_score)
print(best_model)
print(best_model_score)
```

Nous utilisons les meilleurs paramètres déterminés par GridSearchCV dans notre modèle de Machine Learning.

```python
svm_model = SVC(C=0.01, kernel='linear', cache_size=1000)
```

La validation croisé est utilisé afin de déterminer la régularité de performance sur différents échantillon d'entrainement de nos données.

```python
scores = cross_val_score(svm_model, X_train_scaled, y_train_hybrid, cv=5,\
                         scoring='neg_mean_squared_error')
print("Scores de validation croisée:", scores)
```

Cette étape est l'entrainement du modèle sur les données d'entrainement, nous utilisons une méthode de calcul du temps de traitement de cette étape qui nous servira de métrics pour nos modèles de Machine Learning.

```python
start_time = time.time()
svm_model.fit(X_train_scaled, y_train_hybrid)
end_time = time.time()
execution_time = end_time - start_time
print(f"Le temps de traitement est de : {execution_time} secondes")
```

Cette étape est la prédiction de la `TARGET`.

```python
y_pred_hybrid = svm_model.predict(X_test_scaled)
```

Nous étudions le score de notre fonction de scoring permettant d'évaluer la pertinence du modèle de Machine Learning en fonction des FALSE NEGATIVE et des FALSE POSITIVE.

```python
custom_penalty(y_test_hybrid, y_pred_hybrid)
```

Cette métric nous permet d'évaluer différement la pertinence de notre modèle en évaluant le nombre de prédictions correctes en prenant en compte la Précision et le Rappel.

```python
f1 = f1_score(y_test_hybrid, y_pred_hybrid)
print(f"F1-score : {f1}")
```

Cette métric donne le score de pertinence du modèle en prenant en compte les FALSE NEGATIVE et les FALSE POSITIVE mais en appliquant aucune pénalité.

```python
fpr, tpr, thresholds = roc_curve(y_test_hybrid, y_pred_hybrid)
auc_score_hybrid = roc_auc_score(y_test_hybrid, y_pred_hybrid)
auc_score_hybrid
```

Nous utilisons SHAP pour générer un graphique de synthèse qui vous permettra de mieux comprendre l'importance des différentes caractéristiques (features) dans notre modèle prédictif. 

```python
explainer = shap.Explainer(svm_model, X_train_scaled)
shap_values = explainer.shap_values(X_test_scaled)
```

Les points sur l'axe horizontal dans un graphique SHAP de synthèse représentent l'impact de chaque caractéristique sur la sortie du modèle. Les valeurs sur l'axe horizontal correspondent aux valeurs SHAP pour chaque caractéristique. Les valeurs SHAP mesurent la contribution de chaque caractéristique à la différence entre la valeur prédite du modèle pour une observation spécifique et la valeur moyenne des prédictions du modèle sur l'ensemble de données.

```python
feature_names = X_test_hybrid.columns.tolist()
shap.summary_plot(shap_values, X_test_scaled, feature_names=feature_names, show = False)
```

Ce graphique de synthèse SHAP permet d'évaluer globalement l'importance des différentes variables dans la prédiction de notre modèle.

```python
shap.summary_plot(shap_values, X_test_scaled, plot_type="bar", feature_names=feature_names)
```
